```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.ticker as mtick
import matplotlib.pyplot as plt
```

```python
t1=pd.read_csv('WA_Fn-UseC_-Telco-Customer-Churn.csv')
```

```python
t1.head(5)
```

```python
t1.columns.values
```

```python
t1.dtypes
```

```python
#Converting total charges to a numerical type
t1.TotalCharges=pd.to_numeric(t1.TotalCharges,errors='coerce')
t1.isnull().sum()
```

```python
#Removing missing value
t1.dropna(inplace=True)
#print(t1)
#Remove customerId from the dataset
df2=t1.iloc[:,1:]
#print(df2)
#Converting the predictor variable in a binary numeric variable
df2['Churn'].replace(to_replace='Yes',value=1,inplace=True)
df2['Churn'].replace(to_replace='No',value=0,inplace=True)

df_dmy=pd.get_dummies(df2)
df_dmy.head()
```

```python
plt.figure(figsize=(12,10))
df_dmy.corr()['Churn'].sort_values(ascending=False).plot(kind='bar')
```
