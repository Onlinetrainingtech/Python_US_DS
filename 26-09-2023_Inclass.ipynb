```python
#Extracting dependent(input) and independent variables(output)
x=data_set.iloc[:,:-1].values
```

```python
#dependent variable
y=data_set.iloc[:,3]
y
```

```python
from sklearn.impute import SimpleImputer
import numpy as np
```

```python
#impute=impute(miss_value='NaN',strategy='mean',axis=0)
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
```

```python
#Fitting imputer object to the independent variable
imputer1=imputer.fit(x[:,1:3])
```

```python
#Replacing missing data with the calculated mean value
x[:,1:3]=imputer1.transform(x[:,1:3])
```
