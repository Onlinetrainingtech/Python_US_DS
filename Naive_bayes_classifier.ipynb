```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

```python
#1.Collecting data
```

```python
data=pd.read_csv("Datasets/pacific.csv")
data.head()
```

```python
pre_columns=data[:]
pre_columns.drop(['ID','Name','Date','Time','Event','Status','Latitude','Longitude'],axis=1,inplace=True)
pre_columns
```

```python
relevant_cols=pre_columns.columns
relevant_cols
```

```python
train,test=train_test_split(data,test_size=0.3,random_state=5)
train_X=train[relevant_cols]
train_Y=train['Status']
test_X=test[relevant_cols]
test_Y=test['Status']
```

```python
#use the model
gnb=GaussianNB()
```

```python
#Prediction
gnb.fit(train_X,train_Y)
predicted_gnb=gnb.predict(test_X)
```

```python
pd.DataFrame({"Predicated":predicted_gnb,"ActualValue":test_Y}).head()
```
